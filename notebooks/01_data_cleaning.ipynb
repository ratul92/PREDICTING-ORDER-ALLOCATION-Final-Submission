```python
import pandas as pd
import numpy as np
import os

# 1. Define your exact file paths
raw_data_path = r'A:\desktop\New folder\P\Ratul\Supply_Chain_Thesis\data\raw\DataCoSupplyChainDataset.csv'
processed_data_path = r'A:\desktop\New folder\P\Ratul\Supply_Chain_Thesis\data\processed\cleaned_data.csv'

# 2. Load the raw dataset
print("Loading dataset...")
df = pd.read_csv(raw_data_path, encoding='latin1')
print(f"Initial Dataset Shape: {df.shape}")

# 3. Standardize column names
# Converts to lowercase, strips whitespace, and removes parentheses
df.columns = [col.lower().strip().replace(' ', '_').replace('(', '').replace(')', '') for col in df.columns]

# 4. Check for and remove "Impossible Dates"
impossible_mask = df['days_for_shipping_real'] < 0
print(f"Removing {impossible_mask.sum()} records with negative shipping days.")
df = df[~impossible_mask]

# 5. Handle Date Objects (FIXED TYPO HERE)
df['order_date'] = pd.to_datetime(df['order_date_dateorders'])
df['shipping_date'] = pd.to_datetime(df['shipping_date_dateorders'])

# 6. Drop redundant, privacy-risk, or 'future-leaking' columns
cols_to_drop = [
    'order_date_dateorders', 'shipping_date_dateorders', 
    'product_description', 'customer_email', 
    'customer_password', 'customer_fname', 'customer_lname'
]
df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)

print(f"Cleaned Dataset Shape: {df.shape}")

# 7. Save Cleaned Data
os.makedirs(os.path.dirname(processed_data_path), exist_ok=True)
df.to_csv(processed_data_path, index=False)
print(f"Data cleaning complete. File successfully saved to: \n{processed_data_path}")
```
